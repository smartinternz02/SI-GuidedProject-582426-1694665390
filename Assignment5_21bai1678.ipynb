```python
df=pd.read_csv("/content/Mall_Customers.csv")
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

```python
df.head()
```

```python
df['Spending Score (1-100)'].fillna(df['Spending Score (1-100)'].mean(),inplace= True)
df['Annual Income (k$)'].fillna(df['Annual Income (k$)'].mean(),inplace = True)
df['Age'].fillna(df['Age'].mean(),inplace = True)

df['Gender'].fillna(df['Gender'].mode()[0],inplace = True)

```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
df['Gender'] = enc.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
X1=df.drop('Spending Score (1-100)',axis=1)
y1=df['Spending Score (1-100)']
```

```python
X1_train,X1_test,y1_train,y1_test=train_test_split(X1,y1,test_size=0.2,random_state=42)
```

FOR 5 Clusters


FOR 3 Clusters


```python
from sklearn.cluster import KMeans
model3=KMeans(n_clusters=3)
model3.fit(X1_train,y1_train)
```

```python
y3predict=model3.predict(X1_test)
print(y3predict)

```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.array(y3predict)



plt.scatter(y3predict,y1_test,c=data, cmap='viridis', s=50)
plt.colorbar(label='Cluster')

plt.title('Cluster Assignments')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
